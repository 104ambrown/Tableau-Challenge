```python
# importing dependencies
import glob
import pandas as pd
```

```python
#establishing a base path to read in all csvs
files = glob.glob('C:/Users/abrow/Desktop/EMERSON_BOOTCAMP/BOOTCAMP_AMB/Homeworks/HW13_01-29-22_Tableau_Homework/Tableau-challenge/CitiBike_Data/2013*.csv')

#creating an empty list to store the dataframes
frame = []

#writing a simple loops of what to do with the csvs
for file in files:
    frame.append(pd.read_csv(file))

#Concatenate dataframes into 1
citibike2013_df = pd.concat(frame)
```

```python
#Converting starttime and endtime to datetime to prepare for splitting and inserting into new columns
citibike2013_df['Start Date'] = pd.to_datetime(citibike2013_df['starttime']).dt.date
citibike2013_df['Start Time'] = pd.to_datetime(citibike2013_df['starttime']).dt.time
citibike2013_df['Stop Date'] = pd.to_datetime(citibike2013_df['stoptime']).dt.date
citibike2013_df['Stop Time'] = pd.to_datetime(citibike2013_df['stoptime']).dt.time
```

```python
# Writing a function to convert tripduration from seconds to hours, minutes, seconds in a new column
def format_duration(row):
    seconds = row.tripduration
    minutes, seconds = divmod(seconds, 60)
    hours, minutes = divmod(minutes, 60)
    return '{:02d} : {:02d} : {:02d}'.format(hours, minutes, seconds)

# applying the function to the dataframe
citibike2013_df['Trip Duration'] = citibike2013_df.apply(format_duration, axis=1)
citibike2013_df.head()
```
